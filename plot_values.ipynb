```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
epochs = [0, 1, 2, 3, 4]
```

```python
def plot_values(epochs, metric1, metric2, loss1, loss2, metric_name, names):
    plt.rcParams["font.family"] = "serif"
    plt.rcParams["font.size"] = 14
    plt.rcParams["figure.figsize"] = (8,8)
    max_loss = max(1.05, np.max(np.append(loss1, loss2)))
    step = max(0.05, max_loss/20)
    plt.yticks(np.arange(0, max_loss, step))
    
    plt.plot(epochs, metric1, c='tab:blue', label=names[1])
    plt.plot(epochs, metric2, c='tab:gray', label=names[2])    

    plt.plot(epochs, loss1, c='tab:blue', linestyle='dashed')
    plt.plot(epochs, loss2, c='tab:gray', linestyle='dashed')    
    plt.plot([], [], c='black', label=metric_name)
    plt.plot([], [], c='black', linestyle='dashed', label="Loss")   
    
    plt.legend(loc="lower left", prop={'size': 10})
    plt.xlabel("Epochs")
    plt.ylabel(metric_name + " & Loss")
    plt.title(names[0])
    
    plt.show()
```

```python
def plot_values_scales(epochs, metric1, metric2, loss1, loss2, metric_name, names):
    
    plt.rcParams["font.family"] = "serif"
    plt.rcParams["font.size"] = 14
    plt.rcParams["figure.figsize"] = (8,8)
    
    fig, ax1 = plt.subplots()
    plt.title(names[0])
    
    ax1.plot(epochs, loss1, c='tab:blue', linestyle='dashed')
    ax1.plot(epochs, loss2, c='tab:gray', linestyle='dashed')
    ax1.set_xlabel = plt.xlabel("Epochs")
    ax1.set_ylabel("Loss")

    ax2 = ax1.twinx()
    
    ax2.plot(epochs, metric1, c='tab:blue', label=names[1])
    ax2.plot(epochs, metric2, c='tab:gray', label=names[2])
    ax2.plot([], [], c='black', label=metric_name)
    ax2.plot([], [], c='black', linestyle='dashed', label="Loss")  
    ax2.set_ylabel(f"{metric_name}")
    
    plt.legend()
    plt.show()
    

#     max_loss = max(1.05, np.max(np.append(loss1, loss2)))
#     step = max(0.05, max_loss/20)
#     plt.yticks(np.arange(0, max_loss, step))
    
#     plt.plot(epochs, metric1, c='tab:blue', label=names[1])
#     plt.plot(epochs, metric2, c='tab:gray', label=names[2])    

#     plt.plot(epochs, loss1, c='tab:blue', linestyle='dashed')
#     plt.plot(epochs, loss2, c='tab:gray', linestyle='dashed')    
#     plt.plot([], [], c='black', label=metric_name)
#     plt.plot([], [], c='black', linestyle='dashed', label="Loss")   
    
#     plt.legend(loc="lower left", prop={'size': 10})
#     plt.xlabel("Epochs")
#     plt.ylabel(metric_name + " & Loss")
#     plt.title(names[0])
    
#     plt.show()
```
