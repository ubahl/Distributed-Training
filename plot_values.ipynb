```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
epochs = [0, 1, 2, 3, 4]
```

```python
def plot_values(epochs, metric1, metric2, loss1, loss2, metric_name, names):
    plt.rcParams["font.family"] = "serif"
    plt.rcParams["font.size"] = 14
    plt.rcParams["figure.figsize"] = (8,8)
    max_loss = max(1.05, np.max(np.append(loss1, loss2)))
    step = max(0.05, max_loss/20)
    plt.yticks(np.arange(0, max_loss, step))
    
    plt.plot(epochs, metric1, c='tab:blue', label=names[1])
    plt.plot(epochs, metric2, c='tab:gray', label=names[2])    

    plt.plot(epochs, loss1, c='tab:blue', linestyle='dashed')
    plt.plot(epochs, loss2, c='tab:gray', linestyle='dashed')    
    plt.plot([], [], c='black', label=metric_name)
    plt.plot([], [], c='black', linestyle='dashed', label="Loss")   
    
    plt.legend(loc="lower left", prop={'size': 10})
    plt.xlabel("Epochs")
    plt.ylabel(metric_name + " & Loss")
    plt.title(names[0])
    
    plt.show()
```

```python
# one node one GPU
accuracy_001 = [0.7525,0.7875,0.834,0.854,0.8715]
loss_001 = [0.7969508171081543,0.712875247001648,0.5622795224189758,0.5580833554267883,0.5070993304252625]

accuracy_0001 = [0.9305,0.937,0.932,0.9485,0.9455]
loss_0001 = [0.995270848274231,0.4634244740009308,0.3698529601097107,0.28831955790519714,0.2843089997768402]

names = ["One Node, One GPU", "lr=0.001", "lr=0.0001"]
```

```python
plot_values(epochs, accuracy_001, accuracy_0001, loss_001, loss_0001, "Accuracy", names)
```
