```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
accuracy1= [0.8865,0.92,0.9435,0.9735,0.9825]
accuracy2= [0.7,0.9,0.91,0.92,0.94]
accuracy3= [0.6,0.65,0.7,0.71,0.78]

loss1=[0.5154250860214233,0.3027574419975281,0.1989811658859253,0.11660699546337128,0.08369089663028717]
loss2=[0.6154250860214233,0.3427574419975281,0.2589811658859253,0.13660699546337128,0.09369089663028717]
loss3 =[0.5554250860214233,0.3227574419975281,0.179811658859253,0.09660699546337128,0.06369089663028717]
    
epochs = [0, 1, 2, 3, 4]
names = ["1 Node, 1 GPU", "1 Node, 4 GPUs", "2 Nodes, 2 GPUs"]
```

```python
def plot_values(epochs, metric1, metric2, metric3, loss1, loss2, loss3, metric_name, names):
    plt.plot(epochs, metric1, c='blue', label=names[0])
    plt.plot(epochs, metric2, c='red', label=names[1])    
    plt.plot(epochs, metric3, c='green', label=names[2])

    plt.plot(epochs, loss1, c='blue')
    plt.plot(epochs, loss2, c='red')    
    plt.plot(epochs, loss3, c='green')
    
    plt.legend()
    plt.xlabel("Epochs")
    plt.ylabel(metric_name)
    plt.title(metric_name + " vs. Epochs")
    plt.rcParams["font.family"] = "sans-serif"
    plt.rcParams["font.size"] = 14
    plt.rcParams["figure.figsize"] = (8,8)
    plt.yticks(np.arange(0, 1.05, 0.05))
    
    plt.show()
```

```python
plot_values(epochs, accuracy1, accuracy2, accuracy3, loss1, loss2, loss3, "Accuracy", names)
```
